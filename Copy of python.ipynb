```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.amazon.com/s?k=galaxy+compatible+smart+watch"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US"
}
titles=[]
response=requests.get(url,headers=headers)
soup=BeautifulSoup(response.text,"html.parser")
div=soup.find_all("div",class_="sg-col-inner")
for tags in div:
    tag = tags.find_all("h2")
    for t in tag:
      strps=t.text.strip()
      titles.append(strps)
    df=pd.DataFrame(titles,columns=["title"])
    df.to_json("title.json")
    print(strps)
```
