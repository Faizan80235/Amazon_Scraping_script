```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.amazon.com/s?k=galaxy+compatible+smart+watch"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US"
}

titles = []
prices = []
boughtdeatail = []

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

products = soup.find_all("div", class_="sg-col")

for product in products:
    # Title
    title_tag = product.find("h2")
    title = title_tag.text.strip() if title_tag else "No title found"
    titles.append(title)

    # Price
    price_whole = product.find("span", class_="a-price-whole")
    price_fraction = product.find("span", class_="a-price-fraction")
    if price_whole and price_fraction:
        price = price_whole.text.strip() + price_fraction.text.strip()
    else:
        price = "No price found"
    prices.append(price)

    # Bought Detail
    bought = product.find("span", class_="a-size")
    if bought:
        information = bought.text.strip()
    else:
        information = "No Bought found"
    boughtdeatail.append(information)

    print(f"Product: {title}, Price: {price}, Bought: {information}")

# DataFrame
df = pd.DataFrame({
    'title': titles,
    'Price': prices,
    'boughtinformation': boughtdeatail
})

df.to_json("New.json", orient="records", indent=4)
print("✅ Data saved to New.json")
```
